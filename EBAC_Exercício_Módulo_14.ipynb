```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("diamonds")
data.head()
```

```python
diamonds = data[["cut", "price"]].groupby("cut").agg("sum").reset_index()
diamonds.head()
```

```python
grafico = sns.barplot(data=diamonds, x="cut", y="price", ci=None, palette="pastel")
grafico.set(title='Preço do diamante por tipo de corte', xlabel='Tipo de corte', ylabel='Preço do diamante');
```
